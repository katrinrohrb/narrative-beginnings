```python
import gdown 
from pathlib import Path 
import tarfile 

DATA_FOLDER_GDRIVE = "https://drive.google.com/drive/folders/1rGGexqNSwwV2tpIMVct596XabHs1s8UE?usp=sharing"

DATA_FOLDER = Path("../data-test")
```

```python
output = str(DATA_FOLDER
url = DATA_FOLDER_GDRIVE
filenames = gdown.download_folder(url, output = output)
```

```python
for filename in filenames:
    with tarfile.open(filename) as file:
        file.extractall(DATA_FOLDER) 
    Path(filename).unlink()
    
```
